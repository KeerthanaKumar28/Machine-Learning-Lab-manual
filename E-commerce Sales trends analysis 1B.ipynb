# E-commerce sales trend

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data={
    "product_category":np.random.choice(["electronic","clothing","homedecor","books"],100),
    "sales_volume":np.random.randint(10,500,100),
    "price":np.random.uniform(5,500,100).round(2),
    "customer_Age":np.random.randint(18,65,100),
    "purchase_time":pd.date_range(start="2024-01-01",periods=100,freq="D"),
}
```

```python
df=pd.DataFrame(data)
df['month']=df["purchase_time"].dt.to_period("M")
monthly_trends=df.groupby(["month","product_category"])["sales_volume"].sum().unstack(0)
```

```python
monthly_trends.plot(kind="line",figsize=(10,6),marker="o")
plt.title("Montly Sales volume trends by product category")
plt.xlabel("month")
plt.ylabel("sales volume")
plt.legend(title="product category")
plt.grid(True)
plt.show()
```

```python
avg_price=df.groupby("product_category")["price"].mean()
avg_price.plot(kind="bar",figsize=(8,5),color="cyan")
plt.title("Average price by product category")
plt.xlabel("product category")
plt.ylabel("average price")
plt.grid(axis="y")
plt.show()
```
